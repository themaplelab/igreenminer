```python
import Monsoon.HVPM as HVPM
import Monsoon.Operations as op
import Monsoon.sampleEngine as sampleEngine
import subprocess
import time


Mon = HVPM.Monsoon()
Mon.setup_usb()
print('Serial Number:' , Mon.getSerialNumber())
Mon.setVout(3.8)


engine = sampleEngine.SampleEngine(Mon)
engine.enableCSVOutput("./output/InfiniteSampleCount_002.csv")
engine.ConsoleOutput(True)
numSamples= 15*5000 #sampleEngine.triggers.SAMPLECOUNT_INFINITE #sample for one second
test = subprocess.Popen(["adb","shell","sh","/sdcard/test1.sh"], stdout=subprocess.PIPE) #Selecting test case
print('Started Sampling')
engine.startSampling(numSamples)
# adb shell sh /sdcard/test1.sh
# time.sleep(15)
Mon.stopSampling()
print('Disabled sampling!')

engine.disableCSVOutput()
```

```python
import pandas as pd 
import numpy as np

numSamples=1000
#reading csv file 
filepath = '../output/InfiniteSampleCount_002.csv'
df = pd.read_csv(filepath, squeeze = True) 
df = df[['Time(ms)','Main(mA)','Main Voltage(V)']]
df = df.rename(index=str, columns={"Time(ms)": "ms", "Main(mA)": "mA",  "Main Voltage(V)": "v"})
df.head()

df['P'] = df.v * (df.mA / 1000 )
msV = df['ms'].tolist()
msV.insert(0,df['ms'][0])
msV.pop()
len(msV)

df['msOffset'] = msV
df['interval'] = (df['ms']) - (df['msOffset'])
df['joules'] = df['interval'] * df['P']
print('Power:' , np.average(df['P']))
print('Joules:', sum(df['joules']))
df
```
